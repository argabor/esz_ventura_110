A gyufákat karaktergrafikusan ábrázolom.

```python
def match_display(x, n=1):
    """ Karaktergrafikus gyufa rajzolás, 1 sorba """
    
    x = str(x)
    l=lambda s:"".join(s)+"\n"
    h=lambda s:s.replace(*"! ")*~-n+s.replace(*"!_")
    return l(" "+"_ "[c in"14"]*n+" "for c in x)+h(l("| "[c in"237"]+n*"! "[c in"017"]+"| "[c in"156"]for c in x)) \
            +h(l(" |"[c in"01268"]+n*"! "[c in"147"]+"| "[c in"12"]for c in x))

print(match_display(2021))
#print(match_display(1234567890))
```

Mivel sok számjegyből kell hogy álljon, ezért csak hatványozás lehetséges.

```python
def match_display_exp(base, exponent):
    """ Hatványozás ábrázolása """
    
    return '\n'.join([(' '*len(match_display(base).split('\n')[0]) \
                       +'\n'+match_display(base)).split('\n')[:-1][i] + l \
                      for i, l in enumerate(match_display(exponent).split('\n'))])

def check_and_display(base, exponent, msg=None):
    if len(set(f'{base ** exponent}')) == 10:
        print('-'*20)
        print(match_display_exp(base, exponent) + '\n')
        if msg is not None:
            print(msg, end=" ")
        print(f'{base}^{exponent} = {base ** exponent}')
```

Az első 1, vagy 2 számjegy tetejét, ha lebontjuk és áthelyezzük az aljára, akkor kapunk egy hatványt.
 - Csak az első számjegy átalakításához minimum 3 mozgatás szükséges.
 - Az első 2 számjegy átalakításához minimum 5 mozgatás szükséges. Ekkor a kitevő 21 marad.

Ezek a variációk állíthatók elő, amelyek minden számjegyet tartalmaznak:

```python
print(match_display(2021)) #Referencia
check_and_display(4, 821, '3 mozgatás:')
check_and_display(3, 921, '3 mozgatás:')
check_and_display(5, 921, '3 mozgatás:')
check_and_display(3, 621, '3 mozgatás:')
check_and_display(5, 621, '3 mozgatás:')
check_and_display(59, 21, '5 mozgatás:')
check_and_display(39, 21, '5 mozgatás:')
check_and_display(84, 21, '5 mozgatás:')
```

Az utolsó egyesből lehet 11 kitevőt csinálni, úgy, hogy az alső szegmenst a felső mellé helyezzük. Ehhez 1 mozgatás szükséges.
Ekkor a hatvány alapja 202, amit az alábbiak szerint módosíthatunk, maximum 4 mozgatással:

```python
def diffmove_long(old, new):
    """ Meghatározzuk, hogy mennyi gyufát kell hozzáadni és átmozgatni ahhoz,
hogy az első paraméter szerinti számból a második paraméter szerinti számot kapjuk. """
    
    d = (126, 48, 109, 121, 51, 91, 95, 112, 127, 123) # Szegmensek bináris reprezentációja (0-9)
    d_old = ''.join([f'{d[int(c)] :07b}' for c in str(old)]) # Minden gyufát felrakunk, hogy meg tudjuk számolni
    d_new = ''.join([f'{d[int(c)] :07b}' for c in str(new)])
    diff = d_new.count('1') - d_old.count('1') # Ennyit gyufát kell hozzáadni / elvenni
    move = (bin((int(d_old, 2) ^ int(d_new, 2))).count('1') - abs(diff)) // 2 # Ennyit gyufát kell átmozgatni
    return diff, move

def find_all_long(old, t_diffmove):
    """ Az első paraméter szerinti számból, a második paraméterben meghatározott feltétel szerint
átalakítható számokat adja vissza. """
    
    all = set((d for d in range(10**len(str(old))) if diffmove_long(old,d) == t_diffmove))
    return all

base_var_202 = [(f'{1+m} mozgatás', find_all_long(202, (0,m))) for m in range(0,5)]
bulk_base_202 = sum([list(num[1]) for num in base_var_202], [])
bulk_base_202_moves = sum([[num[0] for x in num[1]] for num in base_var_202], [])
```

A lehetséges megoldások, amiben minden számjegy szerepel:

```python
print(match_display(2021)) #Referencia
bases_with_exp_11 = [(num, f'{moves}: {num}^11 = {num**11}') \
                     for num, moves in zip(bulk_base_202, bulk_base_202_moves) if len(set(f'{num**11}')) == 10]

for b in bases_with_exp_11:
    print('-'*20)
    print(match_display_exp(b[0], 11).replace('|', ' ', 2).replace('|  |  \n', '||    \n', 1) + '\n')
    print(b[1])
```
